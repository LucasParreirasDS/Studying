```python
# Importing libraries 
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')

path = '/content/drive/MyDrive/Datasets/'
filename = 'aluguel.csv'

file_path = path + filename

# Loading de dataset
df = pd.read_csv(file_path, sep=';')
df.head(10)
```

### 1. Existem hoteis no dataset? Quais? Onde estão localizados?


```python
# Lets see the uniques values for 'Tipo'
df['Tipo'].value_counts()
```

```python
hotel = df[df['Tipo'] == 'Hotel'].reset_index() # This two records are classified as 'Hotel'
hotel
```

Let's see how other building with +25 rooms are classified

```python
df[df['Quartos'] >= 25]
```

```python
df[df['Tipo'] == 'Apartamento'].sort_values(by='Quartos', ascending=False)
```

```python
print(f"Os hoteis registrados ficam em {hotel['Bairro'][0]} e {hotel['Bairro'][1]}.")
```

### 2. Imoveis com suites estão localizados em quais bairros?

```python
suite = df.groupby('Bairro')['Suites'].count().sort_values(ascending=False)
suite
```

```python
print(f'Imóveis com suítes estão localizados em {len(suite)} bairros diferentes.')
```

### 3. Quais os 10 bairros com condominios mais barato do dataset?

```python
# Os bairros dos 10 condominios mais baratos
cheaper = df[['Bairro', 'Condominio']].sort_values(by='Condominio', ascending=False).head(10)
cheaper[['Bairro']].value_counts()
```

```python
# Os 10 bairros com condominios mais baratos (em media)
cheaper = df.groupby(['Bairro'])['Condominio'].mean().sort_values(ascending=False).head(10)
cheaper
```

### 4. Calcule o IPTU por metro, crie uma coluna com esse resultado

```python
df['IPTU_m2'] = (df.IPTU/df.Area).round(2)
df
```

### 5. Calcule o valor por metro, crie uma coluna com esse resultado

```python
df['Valor_m2'] = (df.Valor/df.Area)
df
```

### 6. Apresente os bairros com o metro mais caro de IPTU

```python
df.groupby('Bairro')['IPTU_m2'].mean().sort_values(ascending=False).head().round(2)
```

### 7. Apresente os bairros com o metro mais caro de Valor

```python
df.groupby('Bairro')['Valor_m2'].mean().sort_values(ascending=False).head().round(2)
```
