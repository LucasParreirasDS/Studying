## 8. Liste todos os filmes de drama



```python
consulta_sql('SELECT * FROM genres')
```

```python
consulta_sql('SELECT m.title, g.name FROM movies AS m INNER JOIN genres AS g ON g.id = m.genre_id = 1 WHERE g.id = 1')
```

## 9. Existem quantos filmes de suspense?

```python
consulta_sql('SELECT g.name AS genre, COUNT(m.title) AS count FROM movies AS m INNER JOIN genres AS g ON g.id = m.genre_id = 1 WHERE g.id = 4')
```
